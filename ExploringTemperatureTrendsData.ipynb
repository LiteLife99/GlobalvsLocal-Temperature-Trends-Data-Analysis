```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the data

```python
pune_df = pd.read_csv('pune_data.csv')
pune_df.head()
```

```python
global_df = pd.read_csv('global_data.csv')
global_df.head()
```

```python
pune_line_plot = pune_df.plot(x="year", y="avg_temp", kind="line")
pune_line_plot.set_xlabel("year")
pune_line_plot.set_ylabel("avg_temp")
pune_line_plot.set_title("Pune temp vs year line plot")
```

```python
pune_df.describe()
```

### Dealing with NULL values and Plotting normal Pune data

```python
#dealing with NULL values
pune_df.info()
#we see that there are 7 NULL values in the avg_temp column
pune_df.mean()
pune_df.fillna(pune_df.mean(), inplace=True)
```

```python
pune_df.info()
#we have succesfully filled the null values with mean of the data
```

```python
pune_line_plot = pune_df.plot(x="year", y="avg_temp", kind="line")
pune_line_plot.set_xlabel("year")
pune_line_plot.set_ylabel("avg_temp")
pune_line_plot.set_title("Pune temp vs year line plot")
```

### Plotting normal global data

```python
global_df.info()
#hence no null values
```

```python
global_line_plot = global_df.plot(x="year", y="avg_temp", kind="line")
global_line_plot.set_xlabel("year")
global_line_plot.set_ylabel("avg_temp")
global_line_plot.set_title("Global temp vs year line plot")
```

```python
global_df.describe()
```

### Performing Moving averages for both data 
Moving averages are used to smooth out data and to make it easier to observe long term trends and not get lost in daily fluctuations.

```python
global_mv_avg = global_df['avg_temp'].rolling(10).mean()
local_mv_avg = pune_df['avg_temp'].rolling(10).mean() #rolling avg of 10 years at a time
```

### Plotting rolling average data to find trend

```python
plt.plot(global_df['year'],global_mv_avg,label='Global')
plt.plot(pune_df['year'],local_mv_avg,label='Pune')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)") 
plt.title("Pune City Average Temperature")
plt.show()
```

## Plotting individual rolling average data to analyse

```python
plt.plot(global_df['year'],global_mv_avg,label='Global')
plt.legend()
plt.xlabel("Years").set_color('white')
plt.ylabel("Temperature (°C)").set_color('white') 
plt.title("Global City Average Temperature").set_color('white')
plt.show()
```

```python
plt.plot(pune_df['year'],local_mv_avg,label='Pune')
plt.legend()
plt.xlabel("Years").set_color('white')
plt.ylabel("Temperature (°C)").set_color('white')
plt.title("Pune City Average Temperature").set_color('white')
plt.show()
```
